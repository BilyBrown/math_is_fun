Let's try a 3 dimensional plot of the equation

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def randrange(n, vmin, vmax):
    """ 
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin
```

```python
m = np.linspace(1, 200, 100)
c = np.linspace(1, 200, 100)
```

```python
M, C = np.meshgrid(m, c)
valid_indices = np.where(M > C)
```

```python
M_valid = M[valid_indices]
C_valid = C[valid_indices]
```

```python
b = (2041.36 - 21 * M_valid - 12 * C_valid) / 12
```

```python
# Now get where b > 0
B_valid_indices = np.where(b > 0)
```

```python
M_valid2 = M_valid[B_valid_indices]
C_valid2 = C_valid[B_valid_indices]
B_valid2 = b[B_valid_indices]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(M_valid2, C_valid2, B_valid2, cmap='viridis')
ax.set_xlabel('m')
ax.set_ylabel('c')
ax.set_zlabel('b')
ax.set_title("Plot of 21m + 12c + 12b = 2041.36")
```

```python
# try for an interactive graph
import plotly.graph_objects as go
# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=M_valid2.flatten(), y=C_valid2.flatten(), z=B_valid2.flatten(),
                                   mode='markers', marker=dict(size=4, color=b.flatten(), colorscale='viridis'))])
# Set labels and title
fig.update_layout(scene=dict(xaxis_title='m', yaxis_title='c', zaxis_title='b'),
                  title='Plot of 21m + 12c + 12b = 2041.36 with constraints m > c > b')
```
