```python
import pandas as pd
import jenkspy
```

```python
df = pd.read_excel(r"C:\Users\bbrown\OneDrive - Green Assets\Desktop\jenks_experiment.xlsx", sheet_name="NaturalJenksTrial")
```

```python
df.head()
```

```python
df.sort_values(by='Value', inplace=True)
```

```python
df['quantile'] = pd.qcut(df['Value'], q=2, labels=['Low', 'High'])
df
```

```python
df['cut_bins'] = pd.cut(df['Value'],
                        bins=2,
                        labels=['Low', 'High'])
df
```

```python
breaks = jenkspy.jenks_breaks(df['Value'],n_classes=2)
breaks
```

```python
df['cut_jenks'] = pd.cut(df['Value'],
                         bins=breaks,
                         labels=['Low', 'High'],
                         include_lowest=True)
df
```

```python
df['cut_jenks4'] = pd.cut(
    df['Value'],
    bins=jenkspy.jenks_breaks(df['Value'], n_classes=3),
    labels=['bucket_1', 'bucket_2', 'bucket_3'],
    include_lowest=True
)
```
